```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pdays181920 = pd.read_pickle("Poltava181920_days_all.pkl")
```

```python
print(pdays181920.columns.values)
```

```python
pdays20_4 = pd.read_pickle("Poltava2020_4_days_all_diagnosis.pkl")
```

```python
print(pdays20_4.columns.values)
```

```python
#pdays181920_HI = pdays181920.groupby('ЄДРПОУ')['Hospital_ID'].agg(list).to_dict()
pdays181920_HI = pdays181920.groupby('ЄДРПОУ')['Hospital'].agg(list).to_dict()
print(pdays181920_HI)
```

```python
Hospitals_181920 = pd.Series(pdays181920['Hospital'].values,index=pdays181920['ЄДРПОУ']).to_dict()
```

```python
n = 0
for k,v in Hospitals_181920.items():
    print(k, v)
    n += 1
    
print(n)
```

```python
Hospitals_20_4 = list(pdays20_4['Hospital_ID'].unique())
```

```python
print(Hospitals_20_4)
print(len(Hospitals_20_4))
```

```python
for h in Hospitals_20_4:
    if h in Hospitals_181920.keys():
        print(h, Hospitals_181920[h])
    else:
        print(h, 'not in Hospitals181920')
```

```python
for k,v in Hospitals_181920.items():
    if k not in Hospitals_20_4:
        print(k, v)
```

```python
Hospitals_181920_20_4 = {}
Hospitals_not_20_4 = {}
for k,v in Hospitals_181920.items():
    if k in Hospitals_20_4:
        Hospitals_181920_20_4[k] = v
    else:
        Hospitals_not_20_4[k] = v
```

```python
print(len(Hospitals_181920_20_4))
print(len(Hospitals_not_20_4))
```

```python
Hospitals_181920_20_4_df = pd.DataFrame.from_dict(Hospitals_181920_20_4, orient = 'index')
Hospitals_181920_20_4_df = Hospitals_181920_20_4_df.reset_index()
#df = df.rename(columns={'$a':'a'})
Hospitals_181920_20_4_df = Hospitals_181920_20_4_df.rename(columns = {'index': 'ЄДРПОУ', 0: 'Лікарня'})


Hospitals_not_20_4_df = pd.DataFrame.from_dict(Hospitals_not_20_4, orient = 'index')
Hospitals_not_20_4_df = Hospitals_not_20_4_df.reset_index()
#df = df.rename(columns={'$a':'a'})
Hospitals_not_20_4_df = Hospitals_not_20_4_df.rename(columns = {'index': 'ЄДРПОУ', 0: 'Лікарня'})

Hospitals = Hospitals_181920_20_4_df.append(Hospitals_not_20_4_df)

print(Hospitals)
```
