```python
Hospitals_Official = pd.read_excel("../data/Poltava_All_Hospitals_Official.xlsx", sheet_name = "Результаты", \
                                   usecols = 'A:G', converters={'Национальный идентификационный номер (ОКПО/НИН)':str})
print(Hospitals_Official)
```

```python
Hospitals_Official = Hospitals_Official.dropna(axis=0, how = 'all')
```

```python
Hospitals_Official
```

```python
Hospitals_Official = Hospitals_Official.loc[:, ['Название компании', 'Национальный идентификационный номер (ОКПО/НИН)', 'Статус']]
```

```python
#res = dict(tuple(d.groupby('a')))
res = dict(tuple(Hospitals_Official.groupby('Национальный идентификационный номер (ОКПО/НИН)')))
```

```python
print(Hospitals_Official.groupby('Национальный идентификационный номер (ОКПО/НИН)').agg(list).to_dict())
```

```python
HO = Hospitals_Official.groupby('Национальный идентификационный номер (ОКПО/НИН)').agg(list).to_dict()
```

```python
Hospitals_Official_dict = {}
for k,v in HO['Название компании'].items():
    Hospitals_Official_dict[k] = [v[0], HO['Статус'][k][0]]
print(Hospitals_Official_dict)
```

```python
a = "123"
print(a[1:])
```

```python
for k,v in Hospitals_Official_dict.items():
    if k not in Hospitals_181920.keys():
        h181920 = Hospitals_181920.get(int(k[1:]), "NaN")
    print(k, v[0], h181920, v[1])
```

```python
for k,v in Hospitals_181920.items():
    if ("0"+str(k)) not in Hospitals_Official_dict.keys():
        if (str(k)) not in Hospitals_Official_dict.keys():
            print(k, v)
```

```python
Hospitals_missing = {}
for k,v in Hospitals_Official_dict.items():
    if int(k) not in Hospitals_181920.keys():
        if v[1] == 'Действующее':
            if 'SANATORII' not in v[0]:
                if 'TOV' not in v[0]:
                    if 'PP' not in v[0]:
                        if 'PRIVATNE'not in v[0]:
                            if 'SANATORNO' not in v[0]:
                                Hospitals_missing[k] = v[0]
```

```python
Hospitals_missing_df = pd.DataFrame.from_dict(Hospitals_missing, orient = 'index')
```

```python
Hospitals_missing_df = Hospitals_missing_df.reset_index()
```

```python
Hospitals_missing_df.to_excel("../mydata/Poltava_Hospitals_Missing.xlsx")
```
